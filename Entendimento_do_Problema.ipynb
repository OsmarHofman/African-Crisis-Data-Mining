#Conjunto de dados sobre tipos de crises em países africanos

````
import pandas as pd 
url = 'https://raw.githubusercontent.com/OsmarHofman/African-Crisis-Data-Mining/master/african_crises.csv?token=AKECPJHSFJDJ77WCEIYKBNC53ZSJ2'
df = pd.read_csv(url,header=0, sep=',')
df
````

##Atributos:
- **case**: Número que representa um país;
- **cc3**: Código de três letras do país;
- **country**: Nome do país;
- **year**: Ano de observação;
- **systemic_crisis**: Indica se ocorreu ou não uma crise sistêmica em determinado ano (0 para não, e 1 para sim);
- **exch_usd**: A taxa de câmbio do país em relação ao dólar;
- **domestic_debt_in_default**: Indica se houve incumprimento da dívida interna soberana ou não em determinado ano (0 para não, e 1 para sim);
- **sovereign_external_debt_default**: Indica se houve incumprimento da dívida externa soberana ou não em determinado ano (0 para não, e 1 para sim);
- **gdp_weighted_default**: Total da dívida inadimplente em relação ao PIB;
- **inflation_annual_cpi**: Taxa anual de inflação do IPC;
- **independence**: Indica se o país é independente ou não (0 para não, e 1 para sim);
- **currency_crises**: Indica se ocorreu ou não uma crise cambial em determinado ano (0 para não, e 1 para sim);
- **inflation_crises**: Indica se ocorreu ou não uma crise de inflação em determinado ano (0 para não, e 1 para sim);
- **banking_crisis**: Indica se ocorreu ou não uma crise bancária em determinado ano (0 para não, e 1 para sim).

##Objetivo:
> O propósito é de tentar prever a **crise sistêmica**(```systemic_crisis```) de determinado país, em um ano, a partir de suas outras informações.
